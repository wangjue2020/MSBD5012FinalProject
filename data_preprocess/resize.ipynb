```python
from PIL import Image
import os, sys
import pandas as pd
```

```python

if not os.path.exists('./resized_test'):
    os.makedirs('./resized_test')
    
```

```python


# 打开文件
path = "./test/"
dest = './resized_test/'
dirs = os.listdir( path )

metadata=pd.read_csv('./test_metadata.csv')
# print(metadata)
size = 256,256

# 输出所有文件和文件夹
for file in dirs:
    if os.path.isdir(file):
        continue
    picture = Image.open(path+file) 
    print("====before")
    print(picture.width,',',picture.height)
#     (width, height) = (picture.width // 2 +1, picture.height // 2 +1)
    picture.thumbnail(size)
    print(picture.width,',',picture.height)
    print("==========")
    metadata.loc[metadata['image_id']==file[:-4],'width'] = picture.width
    metadata.loc[metadata['image_id']==file[:-4],'height'] = picture.height
    picture = picture.save(dest+file)
metadata.to_csv('./resized_test_metadata.csv')


# size = 128, 128

# for infile in glob.glob("*.jpg"):
#     file, ext = os.path.splitext(infile)
#     with Image.open(infile) as im:
#         im.thumbnail(size)
#         im.save(file + ".thumbnail", "JPEG")
```
